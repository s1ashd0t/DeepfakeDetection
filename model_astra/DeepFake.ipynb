```python
import cv2
import os

frame_count = 0
saved_count = 0

def extract_frames(video_path, output_folder, frame_interval=30):
    """
    Extracts frames from a video file and saves them as images.

    Parameters:
    - video_path: Path to the input video file.
    - output_folder: Directory where the frames will be saved.
    - frame_interval: Extract one frame every 'frame_interval' frames.
    """
    os.makedirs(output_folder, exist_ok=True)
    cap = cv2.VideoCapture(video_path)

    global frame_count, saved_count

    while True:
        ret, frame = cap.read()
        if not ret:
            break
        # Save every nth frame based on the interval value
        if frame_count % frame_interval == 0:
            frame_path = os.path.join(output_folder, f"frame_{saved_count:04d}.jpg")
            cv2.imwrite(frame_path, frame)
            saved_count += 1
        frame_count += 1

    cap.release()
    print(f"Extracted {saved_count} frames from {video_path}")

# Example usage:
# video_file = "/content/dataset/Celeb-real/"
# output_dir = "/frames/"

# # extract frames from multiple videos
# for filename in os.listdir(video_file):
#     if filename.endswith(".mp4"):
#         video_path = os.path.join(video_file, filename)

# extract_frames(video_file, output_dir, frame_interval=30)

root, dirs, files = next(iter(os.walk( '/content/dataset/Celeb-real' )))
for f in files :
    input_filename = os.path.join( root, f )
    output_directory = "/content/real/images/"
    seconds = 10
    extract_frames(input_filename, output_directory, seconds)
```

```python
import os
import cv2
import numpy as np
from glob import glob
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.model_selection import train_test_split

# --- PARAMETERS ---
IMG_SIZE = 128
BATCH_SIZE = 32
EPOCHS = 20

# --- DATA LOADING & PREPROCESSING FUNCTIONS ---
def load_and_preprocess_image(img_path):
    # Load image in color (BGR -> convert to RGB)
    image = cv2.imread(img_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    # Resize image
    image = cv2.resize(image, (IMG_SIZE, IMG_SIZE))
    # Normalize to [0, 1]
    image = image.astype("float32") / 255.0
    return image

def prepare_dataset(data_dir, label):
    """
    data_dir: Path to folder containing images (or pre-extracted face crops)
    label: 0 for real images, 1 for deepfake images.
    """
    images = []
    labels = []
    img_files = glob(os.path.join(data_dir, "*.jpg"))  # adjust extension if needed
    for file in img_files:
        try:
            img = load_and_preprocess_image(file)
            images.append(img)
            labels.append(label)
        except Exception as e:
            print(f"Error processing {file}: {e}")
    return np.array(images), np.array(labels)

# Example directory structure (customize these paths as needed)
REAL_IMAGES_DIR = "/content/real/images/"
FAKE_IMAGES_DIR = "/content/real/fake/"

# Load your data (ensure that you have preprocessed the videos into face-cropped images)
X_real, y_real = prepare_dataset(REAL_IMAGES_DIR, label=0)
X_fake, y_fake = prepare_dataset(FAKE_IMAGES_DIR, label=1)

# Combine datasets
X = np.concatenate([X_real, X_fake], axis=0)
y = np.concatenate([y_real, y_fake], axis=0)

# Shuffle and split into train and validation sets (make sure to split by identity if possible)
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, shuffle=True, random_state=42)

# --- DATA AUGMENTATION ---
train_datagen = ImageDataGenerator(
    rotation_range=15,
    horizontal_flip=True,
    zoom_range=0.1,
    brightness_range=[0.8, 1.2]
)
val_datagen = ImageDataGenerator()

train_generator = train_datagen.flow(X_train, y_train, batch_size=BATCH_SIZE)
val_generator = val_datagen.flow(X_val, y_val, batch_size=BATCH_SIZE)

# --- MODEL DEFINITION ---
def build_model(input_shape=(IMG_SIZE, IMG_SIZE, 3)):
    model = Sequential([
        Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=input_shape),
        BatchNormalization(),
        MaxPooling2D((2,2)),
        Dropout(0.25),

        Conv2D(64, (3, 3), activation='relu', padding='same'),
        BatchNormalization(),
        MaxPooling2D((2,2)),
        Dropout(0.25),

        Conv2D(128, (3, 3), activation='relu', padding='same'),
        BatchNormalization(),
        MaxPooling2D((2,2)),
        Dropout(0.3),

        Flatten(),
        Dense(128, activation='relu'),
        Dropout(0.4),
        Dense(1, activation='sigmoid')  # Binary classification: real (0) vs. deepfake (1)
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

model = build_model()
model.summary()

# --- TRAINING ---
history = model.fit(
    train_generator,
    epochs=EPOCHS,
    validation_data=val_generator
)

# --- EVALUATION ---
loss, accuracy = model.evaluate(val_generator)
print(f"Validation Loss: {loss:.4f}, Validation Accuracy: {accuracy:.4f}")
```

```python
import torch
import torch.nn as nn
from efficientnet_pytorch import EfficientNet

class DeepfakeDetector(nn.Module):
    def __init__(self):
        super(DeepfakeDetector, self).__init__()
        self.backbone = EfficientNet.from_pretrained('efficientnet-b0')
        self.backbone._fc = nn.Identity()  # Remove original classifier
        self.fc = nn.Sequential(
            nn.Linear(1280, 256),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(256, 1),
            nn.Sigmoid()
        )

    def forward(self, x):
        features = self.backbone(x)
        out = self.fc(features)
        return out
```

```python
from torch.utils.data import Dataset
import cv2
import albumentations as A
from albumentations.pytorch import ToTensorV2

class DeepfakeDataset(Dataset):
    def __init__(self, image_paths, labels, transform=None):
        self.image_paths = image_paths
        self.labels = labels
        self.transform = transform

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        image = cv2.imread(self.image_paths[idx])
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        if self.transform:
            image = self.transform(image=image)['image']

        label = torch.tensor(self.labels[idx], dtype=torch.float32)
        return image, label
```

```python
from torch.utils.data import DataLoader
import glob
import os

train_transform = A.Compose([
    A.Resize(224, 224),
    A.HorizontalFlip(),
    A.RandomBrightnessContrast(),
    A.Normalize(),
    ToTensorV2(),
])

REAL_IMAGES_DIR = "/content/real/images/"
FAKE_IMAGES_DIR = "/content/real/fake_imag/"

image_paths = glob.glob(os.path.join(REAL_IMAGES_DIR, "*.jpg")) + glob.glob(os.path.join(FAKE_IMAGES_DIR, "*.jpg"))
labels = [0] * len(glob.glob(os.path.join(REAL_IMAGES_DIR, "*.jpg"))) + [1] * len(glob.glob(os.path.join(FAKE_IMAGES_DIR, "*.jpg")))

# Example usage:
train_dataset = DeepfakeDataset(image_paths, labels, transform=train_transform)
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)

```

```python
import torch.optim as optim

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = DeepfakeDetector().to(device)
criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=1e-4)

def train_one_epoch(dataloader):
    model.train()
    running_loss = 0.0

    for images, labels in dataloader:
        images, labels = images.to(device), labels.to(device).unsqueeze(1)
        outputs = model(images)
        loss = criterion(outputs, labels)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    print(f"Loss: {running_loss / len(dataloader):.4f}")

train_one_epoch(train_loader)
```

```python
from sklearn.metrics import accuracy_score

def evaluate(dataloader):
    model.eval()
    preds, truths = [], []

    with torch.no_grad():
        for images, labels in dataloader:
            images = images.to(device)
            outputs = model(images).cpu().numpy()
            preds.extend(outputs > 0.5)
            truths.extend(labels.numpy())

    acc = accuracy_score(truths, preds)
    print(f"Accuracy: {acc:.4f}")


```
